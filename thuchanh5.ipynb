Viết hàm kiểm tra một số có phải là số nguyên tố hay không?

```python
def CheckPrimeNum(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

assert CheckPrimeNum(2) == True
assert CheckPrimeNum(3) == True
assert CheckPrimeNum(4) == False
assert CheckPrimeNum(5) == True

print("All tests passed!")
```

Viết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên.

```python
def PerfectNum(n):
    if n < 2:
        return False
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    return sum == n

assert PerfectNum(6) == True
assert PerfectNum(28) == True
assert PerfectNum(12) == False
assert PerfectNum(496) == True

print("All tests passed!")
```

```python
def float2bin(n) :
    while n > 0:
        n *= 2
        bit = int(n)
        if bit == 1:
            n -= bit
            print('1', end='')
        else:
            print('0', end='')
float2bin(0.625)  # Output: 101
```

Viết hàm trả về tổng các chữ số của số nguyên dương n

```python
def SumDigits(n):
    sum = 0
    while n > 0:
        sum += n % 10
        n //= 10
    return sum
```
