W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về -1

```python
def NumIndex(l, k):
    if k not in l:
        return -1
    return l.index(k)
l = list(map(int, input().split()))
k = int(input())
print(NumIndex(l, k))
```

W6A23 
Viết hàm xóa đi các số nhỏ hơn x có trong mảng

```python
l = list(map(int, input().split()))
x = int(input())

def SortByX(l):
    sorted_list = []
    for i in l:
        if i >= x:
            sorted_list.append(i)
    return sorted_list
print(SortByX(l))
```

W6A24
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
l = list(map(int, input().split()))
def MoveZerosToFirst(l):
    for i in l:
        if i == 0:
            l.pop(i)
    return l
print(MoveZerosToFirst(l))
```

W6A25
Cho dãy chỉ gồm N giá trị từ 0 - 9 . Kiểm tra xem giá trị nào không có trong mảng

```python
l = list(map(int, input().split()))
def SortByN(l):
    OutOfRange = []
    for i in l:
        if i < 0 or i > 9:
            OutOfRange.append(i)
    return OutOfRange
print(SortByN(l))
```

W6A27
Cho một dãy số nguyên, tìm và in ra vị trí các số < x (x là giá trị cho trước), nếu không có in ra -1

```python
l = list(map(int, input().split()))
x = int(input())
def IndexNum(l):
    result = []
    for i in l:
        if i < x:
            result.append(l.index(i))
    return(result if len(result) > 0 else -1)
print(IndexNum(l))
    
```

Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
