Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 2^k hay không?

```python
import math
n = int(input())
if math.log2(n) % 2 == 0:
    print(f"k = {int(log2(n))}") 
else:
    print("Invalid input")
```

Nhập vào số nguyên dương n, tính tổng các chữ số của n

```python
n = int(input())
sum = 0
while n > 0:
    sum += n % 10
    n //= 10
print(sum)
```

Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
odd = 0
even = 0
while n > 0:
    i = n % 10
    if i % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10
print(odd, even)
```

Nhập n
Cho S(k) = 1 + 2 + 3 + … + k
Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n = int(input())
total = 0
k = 0
while k < n-1:
    k += 1
    total += k
print(total)
```
