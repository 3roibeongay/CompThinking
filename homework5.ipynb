Viết hàm kiem tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số nguyên x

```python
def is_perfect_square(x):
    s = int(x**0.5)
    return s * s == x   
# Test cases
assert is_perfect_square(0) == True
assert is_perfect_square(1) == True
assert is_perfect_square(4) == True
assert is_perfect_square(9) == True
assert is_perfect_square(16) == True
print("All tests passed!")
```

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
def detect_number_type(s):
    try:
        int(s)
        return("int")
    except ValueError:
        try:
            float(s)
            return "float"
        except ValueError:
            return "neither"

# Test cases
assert detect_number_type("123") == "int"
assert detect_number_type("-123") == "int"
assert detect_number_type("123.45") == "float"
assert detect_number_type("-123.45") == "float"
assert detect_number_type("abc") == "neither"
print("All tests passed!")  
```

Viết hàm trả về số lượng chữ số khác 0 trong n số nguyên nhập vào từ bàn phím (một list)

```python
def count_nonzero_digits(numbers):
    count = 0
    for num in numbers:
        for digit in str(abs(num)):
            if digit != '0':
                count += 1
    return count

# Test cases
assert count_nonzero_digits([102, 340, 500]) != 4
assert count_nonzero_digits([0, 0, 0]) == 0
assert count_nonzero_digits([-105, 20]) == 3
print("All tests passed!")
```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min]

```python
def max_min(a, b, c):
    return [max(a, b, c), min(a, b, c)]

# Test cases
assert max_min(1, 2, 3) == [3, 1]
assert max_min(-5, 0, 10) == [10, -5]
assert max_min(7, 7, 7) == [7, 7]
print("All tests passed!")
```

Viết hàm tính tổng bình phương các số từ 1 đến n.

```python
def SquaredSum(n):
    sum = 0
    for i in range(1, n + 1):
        sum += i ** 2
    return sum
# Test cases
assert SquaredSum(3) == 14
assert SquaredSum(5) == 55
assert SquaredSum(0) == 0
print("All tests passed!")
```

Viết chương trình tính tổng 2 số được nhập theo biểu thức a + b từ bàn phím (Sử dụng hàm eval)

```python
def SumOfAnB(a, b):
    sum = eval("a" + "+" + "b")
    return sum
a, b = map(int, input("").split())
print(SumOfAnB(a, b))
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau, nếu n ≤ 0 trả về 0

```python
def sum_1_to_n(n):
    if n <= 0:
        return 0

    def has_pairwise_distinct_digits(x):
        s = str(x)
        return len(s) == len(set(s))

    total = 0
    for x in range(1, n + 1):
        if has_pairwise_distinct_digits(x):
            total += x
    return total

# Test cases
assert sum_1_to_n(0) == 0
assert sum_1_to_n(1) == 1
assert sum_1_to_n(10) == 45
assert sum_1_to_n(11) == 45
print("All tests passed!")
```
